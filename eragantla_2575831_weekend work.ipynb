```python
#7.identiying the differences between including added modified deleted lines 
import difflib

# Function to read a file and return its lines as a list
def read_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    return lines

# Input file paths
file1_path = r'C:\Users\phani\OneDrive\Desktop\python folders\source.txt'
file2_path = r'C:\Users\phani\OneDrive\Desktop\python folders\destination.txt'

# Read the contents of the two files
file1_lines = read_file(file1_path)
file2_lines = read_file(file2_path)

# Perform the difference comparison
differ = difflib.Differ()
diff = list(differ.compare(file1_lines, file2_lines))

# Print the differences
print("Differences between file1 and file2:")
for line in diff:
    if line.startswith('- '):
        print(f"Line removed from file1: {line[2:]}")
    elif line.startswith('+ '):
        print(f"Line added to file1: {line[2:]}")
    elif line.startswith('  '):
        print(f"Line unchanged: {line[2:]}")
    elif line.startswith('? '):
        pass  # Skip lines with question marks (used for internal diff info)

# Optionally, you can save the differences to a new file
with open('differences.txt', 'w') as diff_file:
    for line in diff:
        diff_file.write(line + '\n')
```

```python
#8 
# Function to compress a text file
import gzip
def compress_file(input_filename, output_filename):
    with open(input_filename, 'rb') as input_file, gzip.open(output_filename, 'wb') as output_file:
        output_file.writelines(input_file)

# Function to decompress a compressed file
def decompress_file(input_filename, output_filename):
    with gzip.open(input_filename, 'rb') as input_file, open(output_filename, 'wb') as output_file:
        output_file.writelines(input_file)

# Specify the input and output file names
input_filename = r'C:\Users\phani\OneDrive\Desktop\python folders\input.txt'
compressed_filename =  r'C:\Users\phani\Downloads\destination.tar.gzip'
decompressed_filename =r'C:\Users\phani\OneDrive\Desktop\python folders\requirments.txt'

# Compress the text file
compress_file(input_filename, compressed_filename)
print(f'File {input_filename} has been compressed to {compressed_filename}')

# Decompress the compressed file
decompress_file(compressed_filename, decompressed_filename)
print(f'File {compressed_filename} has been decompressed to {decompressed_filename}')
```

```python
#1.1Scenario 
#Adding items to cart
def add_to_cart(cart, product_catalog, product_id, quantity):
    try:
        # Check if the product ID exists in the catalog
        if product_id not in product_catalog:
            raise ValueError("Product ID not found in the catalog")
        
        # Check if quantity is valid (positive integer)
        if not isinstance(quantity, int) or quantity <= 0:
            raise ValueError("Invalid quantity. Please enter a positive integer")
        
        # Add the item to the cart
        cart.add_item(product_id, quantity)
    except ValueError as e:
        print(f"Error: {e}")
```

```python
#1.2.viewing items to cart 
def view_cart(cart):
    try:
        if not cart.has_items():
            raise ValueError("The cart is empty. Add items to your cart first.")
        
        # Display cart contents
        cart.display_contents()
    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
#1.3.proceeding to checout 
def checkout(cart, payment_gateway, stock_manager):
    try:
        # Validate the cart
        cart.validate_cart(stock_manager)
        
        # Process payment
        payment_gateway.process_payment(cart.get_total())
        
        # If payment is successful, finalize the order
        cart.finalize_order()
    except InsufficientStockError as e:
        print(f"Error: {e}. Please remove items from your cart or reduce quantities.")
    except PaymentGatewayError as e:
        print(f"Payment gateway error: {e}. Please try again later or use a different payment method.")
    except PaymentDeclinedError as e:
        print(f"Payment method declined: {e}. Please check your payment details.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        # Ensure resources are released (e.g., close database connections)
        cart.close()
```

```python
#15. Create a custom Python module that includes functions to calculate the factorial of a number and to check if a number is prime. Import and 
def factorial(n):
     
    # single line to find factorial
    return 1 if (n==1 or n==0) else n * factorial(n - 1)
num = 5
print("Factorial of",num,"is",factorial(num))
```
