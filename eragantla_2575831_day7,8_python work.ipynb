```python
#2.BINARY SEARCH EXAMPLE
# Separate function to perform binary search
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid  
        elif arr[mid] < target:
            left = mid + 1  
        else:
            right = mid - 1  
    return -1  
n = int(input("Enter the number of elements in the list: "))
input_list = []
for i in range(n):
    element = int(input(f"Enter element {i + 1}: "))
    input_list.append(element)
input_list.sort()
target = int(input("Enter the target element to search: "))
result = binary_search(input_list, target)
if result != -1:
    print(f"{target} found at position {result}.")
else:
    print(f"{target} is not found in the list.")

```

```python
#6.SINGLY LINKED LIST , LIST CREATING, SMALLEST FROM LIST ,REVERSE ORDER 
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
    def insert_unique(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return

        current = self.head
        while current.next:
            if current.data == data:
                return  
            current = current.next
        if current.data != data:
            current.next = new_node
            
    def find_smallest(self):
        if not self.head:
            return None
        current = self.head
        smallest = current.data
        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next
        return smallest
    
    def display_reverse(self):
        if not self.head:
            return
        stack = []
        current = self.head
        while current:
            stack.append(current.data)
            current = current.next
        while stack:
            print(stack.pop(), end=" -> ")
        print("None")
linked_list = LinkedList()
elements = [23,34,65,78,66]
for element in elements:
    linked_list.insert_unique(element)
smallest = linked_list.find_smallest()
if smallest is not None:
    print(f"The smallest element in the list is: {smallest}")
else:
    print("The list is empty.")
print("Elements in reverse order:")
linked_list.display_reverse()
```

```python
#7.Write a python program to implement the various operations for Stack ADT PUSH POP DISPLAY
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty"

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        if not self.is_empty():
            for item in reversed(self.items):
                print(item)
        else:
            print("Stack is empty")

# Create a stack
stack = Stack()
for i in range (1,11):
    stack.push(i)

while True:
    print("\nStack Operations:")
    print("1. Push")
    print("2. Pop")
    print("3. Display")
    print("4. Quit")
    
    choice = int(input("Enter your choice: "))
    
    if choice == 1:
        item = input("Enter an item to push onto the stack: ")
        stack.push(item)
        print(f"{item} pushed onto the stack.")
    elif choice == 2:
        popped_item = stack.pop()
        print(f"Popped item: {popped_item}")
    elif choice == 3:
        print("Stack contents:")
        stack.display()
    elif choice == 4:
        break
    else:
        print("Invalid choice. Please try again.")
```

```python
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            print("Queue is empty. Cannot dequeue.")

    def display(self):
        if not self.is_empty():
            print("Queue elements:")
            for item in self.items:
                print(item)
        else:
            print("Queue is empty. Nothing to display.")


# Creating a Queue
queue = Queue()
while True:
    print("\nQueue Operations:")
    print("1. Insert (Enqueue)")
    print("2. Delete (Dequeue)")
    print("3. Display")
    print("4. Quit")

    choice = int(input("Enter your choice (1/2/3/4): "))

    if choice == 1:
        item = input("Enter the element to insert: ")
        queue.enqueue(item)
        print(f"{item} has been inserted into the queue.")
    elif choice == 2:
        deleted_item = queue.dequeue()
        if deleted_item is not None:
            print(f"{deleted_item} has been deleted from the queue.")
    elif choice == 3:
        queue.display()
    elif choice == 4:
        break
    else:
        print("Invalid choice. Please enter a valid option.")
```

```python
#9.python to convert infix expression to postfix from push and pop operation stack 
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

def infix_to_postfix(expression):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}
    output = []
    stack = Stack()

    for char in expression:
        if char.isalnum():
            output.append(char)
        elif char == '(':
            stack.push(char)
        elif char == ')':
            while not stack.is_empty() and stack.peek() != '(':
                output.append(stack.pop())
            stack.pop()  
        else:
            while not stack.is_empty() and precedence.get(char, 0) <= precedence.get(stack.peek(), 0):
                output.append(stack.pop())
            stack.push(char)

    while not stack.is_empty():
        output.append(stack.pop())

    return ''.join(output)

# Example a
infix_expr_a = "A/B°C+D*E-F*G"
postfix_a = infix_to_postfix(infix_expr_a)
print("Postfix expression for a:", postfix_a)

# Example b
infix_expr_b = "(B^2-4*A*C)^(1/2) (100)"
postfix_b = infix_to_postfix(infix_expr_b)
print("Postfix expression for b:", postfix_b)
```

```python
#SELECTION SORT ALGORITHM
#3.Write a Python program for sorting a list of elements using selection sort algorithm:

#a.Assume two lists: Sorted list Initially empty and Unsorted List Given input list
#b. In the first iteration, find the smallest element in the unsorted list and place it in the sorted

#c.In the second iteration, find the smallest element in the unsorted list and place it in the correct position by comparing with the element in the sorted list.

#d. In the third iteration, again find the smallest element in the unsorted list and place it in the

#correct position by comparing with the elements in the sorted list.

#e. This process continues till the unsorted list becomes empty

#f. Display the sorted list
def selection_sort(arr):
    
    sorted_list = []  
    while arr:  
        min_value = arr[0]  
        min_index = 0  
        for i in range(len(arr)):
            if arr[i] < min_value: 
                min_value = arr[i]  
                min_index = i 
        sorted_list.append(min_value) 
        arr.pop(min_index)  

    return sorted_list
unsorted_list = [453, 1, 23, 8343, 789]
sorted_result = selection_sort(unsorted_list)
print(" Selection Sort :", sorted_result)
```

```python
#4.INSERTION SORT ALGORITHM
def insertion_sort(arr):
    sorted_list = []
    for unsorted_element in arr:
        if not sorted_list:
            sorted_list.append(unsorted_element)
        else:
            for i in range(len(sorted_list)):
                if unsorted_element < sorted_list[i]:
                    sorted_list.insert(i, unsorted_element)
                    break
                elif i == len(sorted_list) - 1:
                    sorted_list.append(unsorted_element)
    return sorted_list
unsorted_list = [23,45,32,45,11]
sorted_result = insertion_sort(unsorted_list)
print(" insertion Sort :", sorted_result)
        
     
```

```python
#5. MERGE SORT 
def merge_sort(arr):
    if len(arr) <= 1:
        return
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]
    merge_sort(left_half)
    merge_sort(right_half)
    merge(arr, left_half, right_half)

def merge(arr, left, right):
    i = j = k = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1
    while i < len(left):
        arr[k] = left[i]
        i += 1
        k += 1
    while j < len(right):
        arr[k] = right[j]
        j += 1
        k += 1
if __name__ == "__main__":
    numbers = [11,3,4,5,7435,6,7]
    merge_sort(numbers)
    print("Sorted array is:", numbers)
```
